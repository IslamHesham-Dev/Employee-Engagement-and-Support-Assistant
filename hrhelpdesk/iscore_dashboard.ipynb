```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data.csv')
with open('questions.txt', 'r') as file:
    lines = file.readlines()
for line in lines:
    print(line.strip())
```

```python
questions = {}
with open('questions.txt', 'r') as file:
    for line in file:
        number, question = line.strip().split(' - ', 1)
        questions[f'Question {number}'] = [question]

questions = pd.DataFrame(questions)
display(questions)
```

```python
data.head(5)
```

```python
# Transpose the questions DataFrame to make the question numbers columns
questions_T = questions.T
questions_T.columns = ['Question Text']

# Rename the columns in the data DataFrame to match the index of the transposed questions DataFrame
data_renamed = data.rename(columns=questions_T['Question Text'])

# Display the first few rows of the combined DataFrame
display(data_renamed.head())
```

```python
import plotly.express as px

# Get the column name for Question 1 from the renamed data
question_1_column = questions.loc[0, 'Question 1']

# Count the occurrences of each response for Question 1
question_1_counts = data_renamed[question_1_column].value_counts().reset_index()
question_1_counts.columns = ['Response', 'Count']

# Define the desired order of responses
sentiment_order = ['Strongly Agree', 'Agree', 'Neutral', 'Disagree', 'Strongly Disagree']

# Convert the 'Response' column to a categorical type with the specified order
question_1_counts['Response'] = pd.Categorical(question_1_counts['Response'], categories=sentiment_order, ordered=True)

# Sort the DataFrame by the ordered 'Response' column
question_1_counts = question_1_counts.sort_values('Response')

# Define a custom color palette
custom_palette = ['#514384', '#3e4f79', '#cec9db', '#3cc2cb', '#3fbfc2']
# Create a bar chart using Plotly with the custom color palette
fig_q1 = px.bar(question_1_counts, x='Response', y='Count', title=f'Responses to Question 1: {question_1_column}',
                color='Response', color_discrete_sequence=custom_palette) # Map color to Response and use the custom palette

# Display the chart
fig_q1.show()
```

```python
# Get the column name for Question 5 from the renamed data
question_5_column = questions.loc[0, 'Question 5']

# Count the occurrences of each response for Question 5
question_5_counts = data_renamed[question_5_column].value_counts().reset_index()
question_5_counts.columns = ['Response', 'Count']

# Create a pie chart using Plotly
fig_q5 = px.pie(question_5_counts, values='Count', names='Response', color_discrete_sequence=custom_palette, title=f'Responses to : {question_5_column}')

# Display the chart
fig_q5.show()
```

```python
question_7_column = questions.loc[0, 'Question 7']
question_7_counts = data_renamed[question_7_column].value_counts().reset_index()
question_7_counts.columns = ['Response', 'Count']

sentiment_order = ['Strongly Agree', 'Agree', 'Neutral', 'Disagree', 'Strongly Disagree']


# Convert the 'Response' column to a categorical type with the specified order
question_7_counts['Response'] = pd.Categorical(question_7_counts['Response'], categories=sentiment_order, ordered=True)

# Sort the DataFrame by the ordered 'Response' column
question_7_counts = question_7_counts.sort_values('Response')

# Create a treemap using Plotly
fig_q7 = px.treemap(question_7_counts, path=['Response'], values='Count', color_discrete_sequence=custom_palette, title=f'Responses to Question 7: {question_7_column}')

fig_q7.show()
```

```python
import plotly.express as px

# Get the column name for Question 8 from the renamed data
question_8_column = questions.loc[0, 'Question 8']

# Count the occurrences of each response for Question 8
question_8_counts = data_renamed[question_8_column].value_counts().reset_index()
question_8_counts.columns = ['Response', 'Count']

# Define the desired order of responses (assuming similar sentiment scale as Question 1)
sentiment_order = ['Strongly Agree', 'Agree', 'Neutral', 'Disagree', 'Strongly Disagree']


# Convert the 'Response' column to a categorical type with the specified order
question_8_counts['Response'] = pd.Categorical(question_8_counts['Response'], categories=sentiment_order, ordered=True)

# Sort the DataFrame by the ordered 'Response' column
question_8_counts = question_8_counts.sort_values('Response')

# Create a bar chart using Plotly
fig_q8 = px.histogram(question_8_counts, x='Response', y='Count',color= 'Response',color_discrete_sequence=custom_palette, title=f'Responses to Question 8: {question_8_column}')

# Display the chart
fig_q8.show()
```

```python
# Extract the columns related to Question 9 from the renamed data
question_9_data = data_renamed[['Question 9 N1', 'Question 9 N2', 'Question 9 N3', 'Question 9 N4']]

# Rename the columns for Question 9 data
question_9_data = question_9_data.rename(columns={
    'Question 9 N1': 'First Place',
    'Question 9 N2': 'Second Place',
    'Question 9 N3': 'Third Place',
    'Question 9 N4': 'Fourth Place'
})

# Melt the DataFrame to have one row per response per rank
question_9_melted = question_9_data.melt(var_name='Rank', value_name='Characteristic')

# Count the occurrences of each characteristic at each rank
rank_counts = question_9_melted.groupby(['Rank', 'Characteristic']).size().reset_index(name='Count')

# Define the desired order of ranks
rank_order = ['First Place', 'Second Place', 'Third Place', 'Fourth Place']

# Convert the 'Rank' column to a categorical type with the specified order
rank_counts['Rank'] = pd.Categorical(rank_counts['Rank'], categories=rank_order, ordered=True)

# Sort the DataFrame by rank and then by count (descending)
rank_counts = rank_counts.sort_values(['Rank', 'Count'], ascending=[True, False])

# Create a grouped bar chart with the custom color palette
fig_q9 = px.bar(rank_counts, x='Characteristic', y='Count', color='Rank',
             title='Ranking of Leadership Qualities', barmode='group',
             color_discrete_sequence=custom_palette) # Apply the custom color palette

fig_q9.show()
```

```python
import plotly.express as px
import pandas as pd

# Get the column name for Question 10 from the renamed data
question_10_column = questions.loc[0, 'Question 10']

# Count the occurrences of each response for Question 10
question_10_counts = data_renamed[question_10_column].value_counts().reset_index()
question_10_counts.columns = ['Response', 'Count']

# Define the desired order of responses
sentiment_order = ['Strongly Agree', 'Agree', 'Neutral', 'Disagree', 'Strongly Disagree']

# Convert the 'Response' column to a categorical type with the specified order
question_10_counts['Response'] = pd.Categorical(question_10_counts['Response'], categories=sentiment_order, ordered=True)

# Sort the DataFrame by the ordered 'Response' column
question_10_counts = question_10_counts.sort_values('Response')

# Create a bar chart using Plotly
fig_q10 = px.bar(question_10_counts, x='Response', y='Count', title=f'Responses to Question 10: {question_10_column}', color ='Response', color_discrete_sequence=custom_palette)

# Display the chart
fig_q10.show()
```

```python
import plotly.express as px

question_11_column = questions.loc[0, 'Question 11']
question_11_responses = data_renamed[question_11_column].dropna()

# Split responses with multiple selections and stack them
split_responses = question_11_responses.str.split(', ').explode()

# Count the occurrences of each development opportunity
opportunity_counts = split_responses.value_counts().reset_index()
opportunity_counts.columns = ['Opportunity', 'Count']

# Create a pie chart using Plotly
fig_q11 = px.pie(opportunity_counts, values='Count', names='Opportunity', title=f'Responses to Question 11: {question_11_column}', color_discrete_sequence=custom_palette)

fig_q11.show()
```
